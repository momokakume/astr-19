```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
def f(x):
    return a*np.exp(-b*x)
```

```python
a = 2
b = 5
```

```python
y = f(x)
```

```python
def plotting(x, y):
    plt.plot(x,y)
    
    plt.xlabel('x', fontsize=14)
    plt.ylabel('$ae^{-bx}$',fontsize=14)

    plt.savefig('codejournal2_1fig.pdf',bbox_inches='tight')

plotting(x,y)
```
