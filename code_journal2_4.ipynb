```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["font.family"]= "Times New Roman"
plt.rc("font", size=14)
```

```python
x = np.random.binomial(500, 0.5, size=1000)
```

```python
plt.hist(x,bins=100)
plt.xlabel('random number pulled from binomial distribution')
plt.ylabel('number of random numbers pulled')
plt.savefig('code_journal2_4fig.pdf', bbox_inches = 'tight') 
```
