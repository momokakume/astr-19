```python
def f(x):
    return np.sin(x)
```

This first function defines f(x) such that it returns sin(x) for the input x.

```python
def g(x):
    return np.cos(x)
```

This second function defins g(x) such that it returns cos(x) for the input x.

```python
def table():
    x = np.linspace(0,2*np.pi, 1000)
    y = f(x)
    z = g(x)
    
    data = Table([x,y,z], names=['x', 'sin(x)', 'cos(x)'])
    ascii.write(data, 'table.txt', format='commented_header', overwrite=True)
    data_in = ascii.read('table.txt')
    print(data_in)

table()
```
