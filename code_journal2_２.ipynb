```python
#prompt 1

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)

def f(x):
    return a*np.exp(-b*x)

a = 2
b = 5
y = f(x)

def plotting(x, y):
    plt.plot(x,y)

    plt.xlabel('x', fontsize=14)
    plt.ylabel('$ae^{-bx}$',fontsize=14)

    plt.savefig('codejournal2_1fig.pdf',bbox_inches='tight')

plotting(x,y)
```

```python
#prompt 2
plt.rcParams['font.family']= 'Times New Roman'
plt.rc("font", size=18)


sigma = np.linspace(2, 0.2, 100)
scatter = np.random.normal(scale=sigma)
y_err = sigma
y_new = y + scatter

plt.plot(x,y, c = 'teal')
plt.errorbar(x, y_new, yerr=y_err, fmt='.', c = 'palevioletred')
plt.xlabel('x')
plt.ylabel('y (with scatter)')
plt.savefig('codejournal2_2fig.pdf',bbox_inches='tight')
```
