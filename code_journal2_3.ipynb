```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
#prompt 1&2:
plt.rcParams['font.family']= 'Times New Roman'
plt.rc("font", size=18)

x = np.linspace(0, 2*np.pi, 100)

def f(x):
    return a*np.exp(-b*x)
a = 2
b = 5
y = f(x)

sigma = np.linspace(2, 0.2, 100)
scatter = np.random.normal(scale=sigma)
y_err = sigma
y_new = y + scatter

plt.plot(x,y, c='teal')
plt.errorbar(x, y_new, yerr=y_err, fmt='.', c = 'palevioletred')
plt.xlabel('x')
plt.ylabel('y (with scatter)')
```

```python
#prompt 3

def model_line1(x,a,b):
    return a*np.exp(-b*x)

x=np.linspace(0.1, 2*np.pi, 100)

def model_line2(x,a,b):
    return a/(x**b)
```

```python
#model 1
params, params_cov = curve_fit(model_line1, x, y_new, sigma=y_err)

a_fit1 = params[0]
b_fit1 = params[1]

fit_1 = model_line1(x, a_fit1, b_fit1)

print(f'The best fit value of a for first model is {a_fit1}.')
print(f'The best fit value of b for first model is {b_fit1}.')

#model 2
params, prams_cov = curve_fit(model_line2, x, y_new, sigma=y_err)

a_fit2 = params[0]
b_fit2 = params[1]

fit_2 = model_line1(x, a_fit2, b_fit2)

print(f'THe best fit value of a for second model is {a_fit2}.')
print(f'THe best fit value of b for second model is {b_fit1}.')

#plot
plt.errorbar(x,y_new,yerr=y_err, fmt='.', c='palevioletred', label=('data'))
plt.plot(x, fit_1, c='royalblue', label=('best fit line of  $ae^{-bx}$'))
plt.plot(x, fit_2, c='gold', label=('best fit line of  $\dfrac{a}{x^b}$'))

plt.legend(loc="upper right", fontsize='small')

plt.savefig('code_journal2_3fig.pdf', bbox_inches = 'tight')
```
